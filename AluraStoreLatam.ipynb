### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 4. Productos más y menos vendidos

```python
#Tienda 1
import matplotlib.pyplot as plt
ventas = (tienda.groupby('Producto')['Cantidad de cuotas'].sum().sort_values(ascending=False).reset_index())

plt.figure(figsize=(15,5))
bars = plt.bar(ventas['Producto'], ventas['Cantidad de cuotas'])
plt.title('Ventas de los productos de la tienda 1')
plt.xticks(rotation=45, ha='right')
plt.ylabel('')
plt.xlabel('')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
#tienda 2
ventas = tienda2.groupby('Producto')['Cantidad de cuotas'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(15,5)); plt.bar(ventas['Producto'], ventas['Cantidad de cuotas'], color='red')
plt.title('Ventas de los productos de la tienda 2', fontsize=14, weight='bold')
plt.xticks(rotation=45, ha='right')
plt.ylabel('')
plt.xlabel(''); plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout(); plt.show()
```

```python
#t3
# Datos agrupados
ventas = tienda3.groupby('Producto')['Cantidad de cuotas'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(15,5))
plt.bar(ventas['Producto'], ventas['Cantidad de cuotas'], color='green')
plt.title('Ventas de los productos de la tienda 3', fontsize=14, weight='bold')
plt.xticks(rotation=45, ha='right')
plt.ylabel('')
plt.xlabel('')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
#t4
ventas = tienda4.groupby('Producto')['Cantidad de cuotas'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(15,5))
plt.bar(ventas['Producto'], ventas['Cantidad de cuotas'], color='yellow')
plt.title('Ventas de los productos de la tienda 4', fontsize=14, weight='bold')
plt.xticks(rotation=45, ha='right')
plt.ylabel('')
plt.xlabel('')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
